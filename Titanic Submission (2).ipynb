```python
X_trained = trainingdata[['PassengerId', 'Pclass', 'SibSp', 'Parch']]
X_trained.head()
```

```python
Y_trained = trainingdata[['Survived']]
Y_trained.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_trained, Y_trained, test_size=0.2, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()
lgr.fit(X_train, Y_train)
```

```python
x_test = testdata[['PassengerId', 'Pclass', 'SibSp', 'Parch']]
x_test.head()
                  
```

```python
Survivor = lgr.predict(x_test)
Survivor
```

```python
submission = pd.DataFrame({'PassengerId': testdata.PassengerId, 'Survived':Survivor})
submission.head()
```
